```python
import numpy as np
import pandas as pd
from numpy.random import randn
#same set of random num as in video -> using seed(101)
np.random.seed(101)
df = pd.DataFrame(randn(6,4),['A','B','C','D','E','F'],['W','X','Y','Z'])
print(df,type(df),'\n')

#SINGLE COLUMN WE GET SERIES
print(df['W'],type(df['W']),'\n')

#MULTIPLE COLUMNS WE GET DATAFRAME
print(df[['W','Z']],'\n')

#adding new column
df['sum'] =  df['W']+df['X']
print(df)

#deleting col
#axis = 0 refers index, 1 refers col
# for changes to occur use inplace= true
df.drop('sum',axis = 1,inplace= True)
print(df)

#drop index
df.drop('E',inplace = True)
print(df)

#shape to see axis 0 ,1
print(df.shape)

#rows or index
print(df.loc['A'],type(df.loc['A']))
print(df.iloc[2])

#getting values
print(df.loc['A','Y'])

print(df.loc[['A','B'],['W','Y']])
```

```python
#selecting values in entire dataframe

print(df > 0)
booldf = df > 0
print(booldf)
#displays values that are true & Nan for false
print(df[booldf])
print(df[df>0])
```

```python
#selecting values inside a col & displaying df
print(df['W']>0)
#displays only true values NO null values 
print(df[df['W']>0])
print(df[df['Z']<0])
#selecting & displaying particular col
print( df[df['W']>0]['X'] )
```

```python
# same as above more clearly
boolser = df['W']>0
print(boolser)
result = df[boolser]
print(result)
mycols = ['Y','X']
result[mycols]
```

```python
#more than 1 condition
# df[(df['W']>0) and (df['Y']>1)] -> and cant compare series. use &
print(  df[(df['W']>0) & (df['X']>1)]  )
print(  df[(df['W']>0) | (df['X']>1)]  )
```
